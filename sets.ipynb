# Set Operation

# Union (|)
combines all unique elements from both sets.

```python
a1 = {1,2,3,4,5} 
b1 = {4,5,6,7,8} 
c1 = {8,9,10} 
```

```python
a1.union(b1)
```

```python
a1.union(b1,c1)
```

```python
a1 | b1
```

```python
a1|b1|c1
```

# Intersection  (&)
keeps only elements common to both sets.

```python
print(a1)
print(b1)
print(c1) 
```

```python
a1.intersection(b1)
```

```python
a1.intersection(b1,c1)
```

```python
a1&b1
```

# Difference (-)
keeps elements in the first set but not in the second.



```python
a1 = {1,2,3,4,5} 
b1 = {4,5,6,7,8} 
c1 = {8,9,10} 
```

```python
a1.difference(b1)
```

```python
a1 - b1
```

```python
a1 - b1-c1
```

# Symmetric_difference (^)
keeps elements in either set but not both.



```python
a1 = {1,2,3,4,5} 
b1 = {4,5,6,7,8} 
c1 = {8,9,10} 
```

```python
a1 ^ b1
```

# Union update (A |= B or A.update(B)) → adds all elements from B into A.


```python
a1 = {1,2,3,4,5} 
b1 = {4,5,6,7,8} 
c1 = {8,9,10} 
```

```python
a1.update(b1)
```

```python
a1
```

# Intersection update (A &= B or A.intersection_update(B)) → keeps only elements in A that are also in B.


```python
a1 = {1,2,3,4,5} 
b1 = {4,5,6,7,8} 
c1 = {8,9,10} 
```

```python
a1.intersection_update(b1)
```

```python
a1
```

# Difference update (A -= B or A.difference_update(B)) → removes elements from A that are also in B.


```python
a1 = {1,2,3,4,5} 
b1 = {4,5,6,7,8} 
c1 = {8,9,10} 
```

```python
a1.difference_update(b1)
```

```python
a1
```

# Symmetric difference update (A ^= B or A.symmetric_difference_update(B)) → replaces A with elements in either A or B but not both.


```python
a1 = {1,2,3,4,5} 
b1 = {4,5,6,7,8} 
c1 = {8,9,10} 
```

```python
a1.symmetric_difference_update(b1)
```

```python
a1
```

# Superset (>=) → checks if a set contains all elements of another.


```python
s5 = {1,2,3,4,5,6,7,8,9} 
s6 = {3,4,5,6,7,8} 
s7 = {10,20,30,40} 
```

```python
s5.issuperset(s6)
```

```python
s6.issuperset(s5)
```

# Subset (<=) → checks if all elements of one set are in another.


```python
s5 = {1,2,3,4,5,6,7,8,9} 
s6 = {3,4,5,6,7,8} 
s7 = {10,20,30,40} 
```

```python
s6.issubset(s5)
```

```python
s5.issubset(s6)
```

# Disjoint (isdisjoint) → checks if two sets share no elements.


```python
s5 = {1,2,3,4,5,6,7,8,9} 
s6 = {3,4,5,6,7,8} 
s7 = {10,20,30,40} 
```

```python
s5.isdisjoint(s6)
```

```python
s6.isdisjoint(s5)
```

```python
s7.isdisjoint(s5)
```
