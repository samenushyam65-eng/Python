# Indexing

```python
str = 'Hello World'
str
```

```python
str[0]   # it will return the first character of the string
```

```python
print(str[-1])   # it will return the last character of the string
print(str[len(str)-1])   # it will return the last character of the string
```

# Key Rule
When step is negative, Python expects start > stop.
Here we have start = 1 and stop = 8. Since 1 < 8, the slice is invalid in direction.


```python
s[1:8:-2]  
```

```python
s[8:1:-2]
```

```python
s[8:1:-3]
```
