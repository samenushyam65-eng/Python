# 1. Case Conversion

```python
text = "Python is a high-level" 
```

```python
text.upper()    # it will convert the string to upper case
```

```python
text.lower()    # it will convert the string to lower case
```

```python
text.title()    # it will convert the first letter of each word to upper case
```

```python
text.capitalize()    # it will convert the first letter of the string to upper case and the rest to lower case
```

```python
text.swapcase()    # it will convert the upper case to lower case and lower case to upper case
```

# 2. Searching & Checking

```python
text
```

```python
text.find("is")    # it will return the index of the first occurrence of the substring "is"
```

```python
text.find('off')    # it will return -1 if the substring is not found
```

```python
print(text.index('is'))    # it will return the index of the first occurrence of the substring "is"
print(text.index('off'))    # it will raise a ValueError if the substring is not found
```

```python
print(text.startswith("Python"))    # it will return True if the string starts with "Python"
print(text.startswith("python"))    # it will return False because it is case sensitive
```

```python
print(text.endswith('level'))    # it will return True if the string ends with "level"
print(text.endswith('Level'))    # it will return False because it is case sensitive
```

```python
text.count('i')    # it will return the number of occurrences of the substring 'i' in the string
```

```python
text1 = "Hello123"
```

```python
print(text1.isalnum())    # it will return True if all characters in the string are alphanumeric (letters and numbers) and there is at least one character, otherwise it will return False
print(text1.isalpha())    # it will return True if all characters in the string are alphabetic (letters) and there is at least one character, otherwise it will return False
print(text1.isdigit())    # it will return True if all characters in the string are digits and there is at least one character, otherwise it will return False
```

# 4. Formatting

```python
a= 3
b=4
c=a+b
print(f"{a}+{b}={c}")    # it will print the string with the values of a, b and c
print("{}+{}={}".format(a,b,c))    # it will print the string with the values of a, b and c
```

```python
text = 'Hello World'
print(text.ljust(20))    # it will left justify the string in a field of width 20
print(text.center(20))    # it will center the string in a field of width 20
print(text.rjust(20))    # it will right justify the string in a field of width
```

```python
print(text.ljust(20, '-'))    # it will left justify the string in a field of width 20 and fill the empty space with '-'
print(text.center(20, '-'))    # it will center the string in a field of width 20 and fill the empty space with '-'
print(text.rjust(20, '-'))    # it will right justify the string in a field of width 20 and fill the empty space with '-'
```
