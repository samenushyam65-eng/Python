# Copy Dictionary

```python
mydict1 = mydict    # create a reference of mydict
```

```python
mydict1
```

```python
mydict['Address'] = 'Delhi' # if we do changes in the mydict then it will impact on mydict1 also
print(mydict)
print(mydict1)
```

```python
id(mydict1),id(mydict)
```

```python
mydict2=mydict.copy()      # create a copy of the dictionary
```

```python
mydict['Address'] = 'Delhi' # if we do changes in the mydict then it does not impact on mydict1
print(mydict)
print(mydict2)
```

```python
id(mydict),id(mydict2)
```

# Looping Through a Dictionay

```python
mydict1 = {'Name':'Asif' , 'ID': 12345 , 'DOB': 1991 , 'Address' : 'Hilsinki' ,'Job':'Analyst'} 
mydict1
```

```python
for i in mydict1:
    print(i)
```

```python
for i in mydict1:
    print(mydict1[i])
```

```python
for i in mydict1:
    print(f"{i} : {mydict1[i]}")
```

# Dictionary membership 

```python
mydict1 = {'Name':'Asif' , 'ID': 12345 , 'DOB': 1991 , 'Job': 'Analyst'} 
mydict1
```

```python
'Name' in mydict1  # Test if a key is in a dictionary or not.
```

```python
'Asif' in mydict1  # Membership test can be only done for keys.
```

```python
all(mydict1)     # all() returns True - If all all keys of the dictionary are true else false
```

```python
any(mydict1)  # any() returns True if any key of the dictionary is True. If not, any() returns False.
```
