```python
a=3
b=4
c=a+b
```

```python
print(f"The sum of {a} and {b} is {c}.")
print("The sum of {} and {} is {}".format(a,b,c))
print("The sum of %d and %d is %d" % (a,b,c))
print("The sum of",a,"and",b,"is",c)
```

```python
print(a,end=" ")     # By default, the end parameter is set to a newline character (\n), which means that each print statement will be printed on a new line. By setting end to a space (" "), we can print multiple values on the same line with a space in between.
print(b,end=" ")
print(c)
```

```python
print(a,b,c,sep="-")   # The sep parameter is used to specify the separator between the values being printed. By default, it is set to a space (" "), but we can change it to any string we want. In this case, we set it to a hyphen ("-"), so the output will be "3-4-7".
```
